# Data Modeling

## Question

- Design and implement a data model and relational database schema for a simple travel management application.


- The conceptual definition of the model is below. You must:

    - Use the MySQL or other tools to produce an E-R diagram.


- The entities are:
    - _Customer_
    - _TravelProvider_
    - _Itinerary_
    
    
- A _Customer_ has the following information:
    - ID
    - last_name
    - first_name
    - email
    - zero or more itineraries (booked trips)
        
        
- A _TravelProvider_ has the following information:
    - ID

- An _Itinerary_ has the following information:
    - ID

        - A hotel state has the following properties:
            - ID
            - Travel provider ID
            - The customer for whom this ReservationItem applies.
            - Start date
            - End date
            - Kind is one of _single,_ _king,_ _suite._
            - Rate is the rate per day.
        - A flight has the following properties:
            - ID
            - Travel provider ID
            - The customer for whom this ReservationItem applies.
            - Date
            - Time
            - Departure airport code (from https://en.wikipedia.org/wiki/IATA_airport_code)
            - Arrival airport code (from https://en.wikipedia.org/wiki/IATA_airport_code)
            - The flight number uniquely determines the airport codes, departure time and duration.

````python
%%sql

CREATE TABLE `Customer` (
    `ID` int(11) NOT NULL,
    `last_name` varchar(64) NOT NULL,
    `first_name` varchar(64) NOT NULL,
    `email` varchar(128) NOT NULL,
    PRIMARY KEY (`ID`),
    UNIQUE KEY `email_UNIQUE` (`email`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;


CREATE TABLE `Itinerary` (
    `ID` int(11) NOT NULL,
    PRIMARY KEY (`ID`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;


CREATE TABLE `TravelProvider` (
    `ID` int(11) NOT NULL,
    `name` varchar(64) NOT NULL,
    `kind` enum('AL', 'HT') NOT NULL,
    PRIMARY KEY (`ID`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;


CREATE TABLE `Airports` (
    `code` varchar(4) NOT NULL,
    `name` varchar(64) DEFAULT NULL,
    PRIMARY KEY (`code`)
) ENGINE=InnoDB DEFAULT CHARSET=utf8;


CREATE TABLE `Flight` (
    `flight_no` varchar(12) NOT NULL,
    `airline` int(11) NOT NULL,
    `departure_airport` varchar(4) NOT NULL,
    `arrival_airport` varchar(4) NOT NULL,
    `departure_time` time NOT NULL,
    `arrival_time` time NOT NULL,
    `duration` float NOT NULL,
    PRIMARY KEY (`flight_no`),
    KEY `airline_fk_idx` (`airline`),
    KEY `dept_code_idx` (`departure_airport`),
    KEY `arriv_code_idx` (`arrival_airport`),
    CONSTRAINT `airline_fk` FOREIGN KEY (`airline`) REFERENCES `TravelProvider` (`ID`) ON DELETE NO ACTION ON UPDATE NO ACTION,
    CONSTRAINT `arriv_code` FOREIGN KEY (`arrival_airport`) REFERENCES `Airports` (`code`) ON DELETE NO ACTION ON UPDATE NO ACTION,
    CONSTRAINT `dept_code` FOREIGN KEY (`departure_airport`) REFERENCES `Airports` (`code`) ON DELETE NO ACTION ON UPDATE NO ACTION
) ENGINE=InnoDB DEFAULT CHARSET=utf8;


CREATE TABLE `ItineraryCustomer` (
    `customer_id` int(11) NOT NULL,
    `itinerary_id` int(11) NOT NULL,
    `type` enum('P', 'S') NOT NULL,
    PRIMARY KEY (`customer_id`, `itinerary_id`, `type`),
    KEY `itinerary_fk_idx` (`itinerary_id`),
    CONSTRAINT `i_to_c` FOREIGN KEY (`customer_id`) REFERENCES `Customer` (`ID`) ON DELETE NO ACTION ON UPDATE NO ACTION,
    CONSTRAINT `i_to_i` FOREIGN KEY (`itinerary_id`) REFERENCES `Itinerary` (`ID`) ON DELETE NO ACTION ON UPDATE NO ACTION
) ENGINE=InnoDB DEFAULT CHARSET=utf8;


CREATE TABLE `ItineraryItem` (
    `ID` int(11) NOT NULL,
    `itinerary_id` int(11) NOT NULL,
    `customer_id` int(11) NOT NULL,
    `travel_provider` int(11) NOT NULL,
    `reservation_kind` enum('AL', 'HT') NOT NULL,
    PRIMARY KEY (`ID`),
    KEY `itinerary_fk_idx` (`itinerary_id`),
    KEY `provider_fk_idx` (`travel_provider`),
    KEY `i_to_c_idx` (`customer_id`),
    CONSTRAINT `ii_to_cc` FOREIGN KEY (`customer_id`) REFERENCES `Customer` (`ID`) ON DELETE NO ACTION ON UPDATE NO ACTION,
    CONSTRAINT `itinerary_fk` FOREIGN KEY (`itinerary_id`) REFERENCES `Itinerary` (`ID`) ON DELETE NO ACTION ON UPDATE NO ACTION,
    CONSTRAINT `provider_fk` FOREIGN KEY (`travel_provider`) REFERENCES `TravelProvider` (`ID`) ON DELETE NO ACTION ON UPDATE NO ACTION
) ENGINE=InnoDB DEFAULT CHARSET=utf8;


CREATE TABLE `FlightReservation` (
    `ID` int(11) NOT NULL,
    `flight_no` varchar(11) NOT NULL,
    `date` date NOT NULL,
    `itinerary_item_id` int(11) NOT NULL,
    PRIMARY KEY (`ID`),
    KEY `item_fk_idx` (`itinerary_item_id`, `ID`),
    KEY `flight_fk_idx` (`flight_no`),
    CONSTRAINT `item_fk` FOREIGN KEY (`itinerary_item_id`) REFERENCES `ItineraryItem` (`ID`) ON DELETE NO ACTION ON UPDATE NO ACTION,
    CONSTRAINT `flight_fk` FOREIGN KEY (`flight_no`) REFERENCES `Flight` (`flight_no`) ON DELETE NO ACTION ON UPDATE NO ACTION
) ENGINE=InnoDB DEFAULT CHARSET=utf8;


CREATE TABLE `HotelReservation` (
    `ID` int(11) NOT NULL,
    `start_date` date NOT NULL,
    `end_date` date NOT NULL,
    `kind` enum('K', 'S', 'D') NOT NULL,
    `rate` float NOT NULL,
    `itinerary_item_od` int(11) NOT NULL,
    PRIMARY KEY (`ID`),
    KEY `itinerary_item_id_idx` (`itinerary_item_od`),
    CONSTRAINT `itinerary_item_id` FOREIGN KEY (`itinerary_item_od`) REFERENCES `ItineraryItem` (`ID`) ON DELETE NO ACTION ON UPDATE NO ACTION
) ENGINE=InnoDB DEFAULT CHARSET=utf8;
````

# Query Question I

## Question

- The career summary integrates information from Lahman2017. The relevant tables are:
    - People
    - Appearances
    - Batting
    - Pitching
    - Fielding
    
    
- A sample of running the query is below.


- Columns:
    - _Games_ is $G\_all$ from _Appeareances._
    - Batting: _Hits_ is sum of _H,_ _At Bats_ is sum of _AB,_ _RBIs_ is sum of _RBI._ _Average_ is the sum of _H_ divided by the sum of _AB._
    - Pitching: $P\_wins$ is the sum of _W_ and $P\_loss$ is the sum of _L._
    - Fielding: _Put Outs_ is sum of _PO,_ _Assists_ is sum of _A_ and _Errors_ is some of _E._
    
    
- All numeric types must be integers, except for _Average._ You must round _Average_ to 3 decimal places.


- Your column names must match.


- You should be able to figure out what "--" means and how to produce it.


## Answer

````python
%%sql

SELECT 
        `e`.`playerID` AS `playerID`,
        `e`.`teamID` AS `teamID`,
        `e`.`yearID` AS `yearID`,
        `e`.`Games` AS `Games`,
        `e`.`Hits` AS `Hits`,
        `e`.`At Bats` AS `At Bats`,
        `e`.`average` AS `average`,
        `e`.`HRs` AS `HRs`,
        `e`.`RBIs` AS `RBIs`,
        `e`.`P_wins` AS `P_wins`,
        `e`.`P_loss` AS `P_loss`,
        `f`.`Put Outs` AS `Put Outs`,
        `f`.`Assists` AS `Assists`,
        `f`.`Errors` AS `Errors`
    from
        (((SELECT
        `c`.`playerID` AS `playerID`,
            `c`.`teamID` AS `teamID`,
            `c`.`yearID` AS `yearID`,
            `c`.`Games` AS `Games`,
            `c`.`Hits` AS `Hits`,
            `c`.`At Bats` AS `At Bats`,
            `c`.`average` AS `average`,
            `c`.`HRs` AS `HRs`,
            `c`.`RBIs` AS `RBIs`,
            `d`.`P_wins` AS `P_wins`,
            `d`.`P_loss` AS `P_loss`
        from
            (((SELECT
            `a`.`playerID` AS `playerID`,
                `a`.`teamID` AS `teamID`,
                `a`.`yearID` AS `yearID`,
                `a`.`Games` AS `Games`,
                `b`.`Hits` AS `Hits`,
                `b`.`At Bats` AS `At Bats`,
                ifnull(`b`.`average`, "--") AS `average`,
                `b`.`HRs` AS `HRs`,
                `b`.`RBIs` AS `RBIs`
        from
            (((SELECT
            `lahman2017`.`Appearances`.`playerID` AS `playerID`,
                `lahman2017`.`Appearances`.`teamID` AS `teamID`,
                `lahman2017`.`Appearances`.`yearID` AS `yearID`,
                `lahman2017`.`Appearances`.`G_all` AS `Games`
        from
            `lahman2017`.`Appearances`)) `a`
        LEFT JOIN (SELECT
            `lahman2017`.`Batting`.`playerID` AS `playerID`,
                `lahman2017`.`Batting`.`teamID` AS `teamID`,
                `lahman2017`.`Batting`.`yearID` AS `yearID`,
                CAST(SUM(`lahman2017`.`Batting`.`H`) AS UNSIGNED) AS `Hits`,
                CAST(SUM(`lahman2017`.`Batting`.`AB`) AS UNSIGNED) AS `At Bats`,
                CAST(SUM(`lahman2017`.`Batting`.`H`)/SUM(`lahman2017`.`Batting`.`AB`) AS decimal(10, 3)) AS `average`,
                CAST(SUM(`lahman2017`.`Batting`.`HR`) AS UNSIGNED) AS `HRs`,
                CAST(SUM(`lahman2017`.`Batting`.`RBI`) AS UNSIGNED) AS `RBIs`
        from
            `lahman2017`.`Batting`
        GROUP BY `lahman2017`.`Batting`.`playerID`, `lahman2017`.`Batting`.`teamID`, `lahman2017`.`Batting`.`yearID`) `b` ON (((`a`.`playerID` = `b`.`playerID`)
            and (`a`.`teamID` = `b`.`teamID`)
            and (`a`.`yearID` = `b`.`yearID`)))))) `c`
        LEFT JOIN (SELECT
            `lahman2017`.`Pitching`.`playerID` AS `playerID`,
                `lahman2017`.`Pitching`.`teamID` AS `teamID`,
                `lahman2017`.`Pitching`.`yearID` AS `yearID`,
                CAST(SUM(`lahman2017`.`Pitching`.`W`) AS UNSIGNED) AS `P_Wins`,
                CAST(SUM(`lahman2017`.`Pitching`.`L`) AS UNSIGNED) AS `P_loss`
        from 
            `lahman2017`.`Pitching`
        GROUP BY `lahman2017`.`Pitching`.`playerID`, `lahman2017`.`Pitching`.`teamID`, `lahman2017`.`Pitching`.`yearID`) `d` ON (((`c`.`playerID` = `d`.`playerID`)
            and (`c`.`teamID` = `d`.`teamID`)
            and (`c`.`yearID` = `d`.`yearID`)))))) `e`
        LEFT JOIN (SELECT
            `lahman2017`.`Fielding`.`playerID` AS `playerID`,
                `lahman2017`.`Fielding`.`teamID` AS `teamID`,
                `lahman2017`.`Fielding`.`yearID` AS `yearID`,
                CAST(SUM(`lahman2017`.`Fielding`.`PO`) AS UNSIGNED) AS `Put Outs`,
                CAST(SUM(`lahman2017`.`Fielding`.`A`) AS UNSIGNED) AS `Assists`,
                CAST(SUM(`lahman2017`.`Fielding`.`E`) AS UNSIGNED) AS `Errors`
        from
            `lahman2017`.`Fielding`
        GROUP BY `lahman2017`.`Fielding`.`playerID`, `lahman2017`.`Fielding`.`teamID`, `lahman2017`.`Fielding`.`yearID`) `f` ON (((`e`.`playerID` = `f`.`playerID`)
            and (`e`.`teamID` = `f`.`teamID`)
            and (`e`.`yearID` = `f`.`yearID`))))
limit 20;
````

# Query Question II

## Question

- In the simple version of Moneyball, we tried to predict a team's wins based on a team's performance statistics.


- A simple sample of the initial data we loaded is:

````python
%%sql

select teamid, yearid, w, ab, h, `2b` as b2, `3b` as b3, BB, hr, era
    from lahman2017.teams
    where yearid >= 2000
    order by yearid asc
    limit 10;
````

- The Boston Red Sox hit 167 homeruns (hr) in 2000. Was 167 homeruns good or bad? You cannot answer the question in isolation. The answer depends on the homerun production of other teams.


- Normalizing performance require two steps:
    1. Computing average and aggregate behavior in a season over all teams.
    2. Understanding how a team's metrics compared to aggregates and averages for the season.
    
    
- Write __a single query__ that produces a relative performance table per team, per year.
    - For w, ab, h, b2, b3, hr, and bb, the corresponding column is the team's value divided by the total over all teams for the year.
    - For ERA, the value is the team's ERA relative to the average ERA over all teams for the year.

## Answer

````python
%%sql
SELECT 
    teamid, yearid,
    round(t.w/average.w, 3) as w,
    round(t.ab/average.ab, 3) as ab,
    round(t.h/average.h, 3) as h,
    round(t.b2/average.b2, 3) as b2,
    round(t.b3/average.b3, 3) as b3,
    round(t.hr/average.hr, 3) as hr,
    round(t.bb/average.bb, 3) as bb,
    round(t.era/average.era, 3) as real_era
from
    (SELECT
        yearid, sum(w) as w, sum(ab) as ab,
        sum(h) as h, sum(`2b`) as b2, sum(`3b`) as b3, sum(BB) as BB,
        sum(hr) as hr, round(avg(era), 3) as era
    from lahman2017.teams where yearid >= 2000
    group by yearid) as average
join
    (SELECT teamid, yearid, w, ab, h, `2b` as b2, `3b` as b3, BB, hr, era
    from lahman2017.teams where yearid >= 2000) as t
using (yearid)
limit 20;
````
